```python
# 读取数据
df = pd.read_excel('ML_SACCVT_NOT_DEDUCT_CL_TP_FGS.xlsx')

location_ls = df['三者车修理厂'].values.tolist()
```

```python
locations = ['安徽','常州','江苏','苏州','温州','无锡','浙江','宁波','上海']
functions = {0: anhui, 1: changzhou, 2: jiangsu, 3: suzhou, 4: wenzhou, 5: wuxi, 6: zhejiang, 7: ningbo, 8: shanghai}
 
exec_outcome = []
for k,loc in enumerate(location_ls):
    flag = 0
    # 空值用“通赔分公司”填充
    if df['三者车修理厂'].isnull().iloc[k] == True:
        exec_outcome.append(df['通赔分公司'].iloc[k].replace('分公司',''))
        
    else:
        for i in range(len(functions)):
            func = functions[i]
            flag = func(location_ls[k])
            if flag == 1:
                exec_outcome.append(locations[i])
                break
        
        # 至于无法匹配的数据用“通赔分公司”填充
        if flag == 0:
            exec_outcome.append(df['通赔分公司'].iloc[k].replace('分公司',''))
```

```python
df['修理厂所在地'] = exec_outcome
df.to_excel('ML_SACCVT_NOT_DEDUCT_CL_TP_FGS.xlsx',encoding='utf-8',index=0)
```
